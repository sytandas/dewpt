```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('assets/Indian_Climate_Dataset_2024_2025.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
# TODO: finding the city wise dew point and graph, which city best for running weather from this data.

# dew point
# ── 2. Derive Dew Point (Magnus formula) ─────────────────────
# Td = (b*alpha) / (a - alpha), alpha = ln(RH/100) + (a*T)/(b+T)
a, b = 17.625, 243.04
T = df['Temperature_Avg (°C)']
RH = df['Humidity (%)']
alpha = np.log(RH / 100.0) + (a * T) / (b + T)
x = df['DewPoint'] = (b * alpha) / (a - alpha)
```
